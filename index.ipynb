## Question 2

Explain the relationship between **Primary Keys** and **Foreign Keys**.

Type your answer below this line:
_______________________________________________________________________________________________________________________________
Primary keys are the values contained in a table column that uniquely identify each record in that table. A table cannot have two primary keys with the same value. A foreign key is a value in another table that refers to a primary key in a different table. 




## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:

one to one: each record in the first table refers to only one record in the second table

One to many: Each record in the first database is found multiple times in a second database (but each record in the second database relates to only one record in the first database)

many to many: a value in the first table can refer to multiple values in the second table, and vice versa.
